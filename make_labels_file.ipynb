```python
import pandas as pd
import twd97

# Read the CSV files into pandas DataFrames
file_path = "D:\\Yehmh\\test_py\\belt_DBH_5CMup.xlsx"
df = pd.read_excel(file_path)

# Group species data by trap location and find the most common species in each trap
main_species_df = df.groupby(['Plot', 'Transect', 'Quadrat'])['CSP'].agg(lambda x: x.value_counts().idxmax()).reset_index()
# print(main_species_df)

# Calculate the counts of each species in each plot
species_counts = df.groupby(['Plot', 'Transect', 'Quadrat', 'CSP']).size().reset_index(name='Count')
# print(species_counts)

# Calculate the total count of species in each plot
total_counts = species_counts.groupby(['Plot', 'Transect', 'Quadrat'])['Count'].sum().reset_index(name='TotalCount')
# print(total_counts)

# Merge total counts with species counts
species_counts = pd.merge(species_counts, total_counts, on=['Plot', 'Transect', 'Quadrat'], how='left')
# print(species_counts)

# # Calculate the percentage of the main species in each plot
species_counts['Percentage'] = (species_counts['Count'] / species_counts['TotalCount']) * 100
species_80up = species_counts[species_counts['Percentage'] >= 80]
print(species_80up)
# print(species_counts)

# # Now, you can save the merged DataFrame to a new CSV file if needed
# species_counts.to_csv("belt_species_percentage.csv", index=False)
species_80up.to_excel("species_80up.xlsx", index = False)
```

```python
import pandas as pd

# Load the Excel file
df = pd.read_excel("D:\\Yehmh\\test_py\\belt_location.xlsx")

# Function to calculate missing coordinates
def fill_coordinates(row):
    if pd.notna(row['latitude_lu']):
        return row['latitude_lu'], row['longitude_lu']
    elif pd.notna(row['latitude_ld']):
        return row['latitude_ld'] + 20 * 0.000008983, row['longitude_ld']
    elif pd.notna(row['latitude_ru']):
        return row['latitude_ru'], row['longitude_ru'] - 100 * 0.000008983
    elif pd.notna(row['latitude_rd']):
        return row['latitude_rd'] + 20 * 0.000008983, row['longitude_rd'] - 100 * 0.000008983

# Apply the function to fill missing coordinates
df['latitude_lu'], df['longitude_lu'] = zip(*df.apply(fill_coordinates, axis=1))

# Display the updated dataframe
print(df)

# Save the updated dataframe to CSV
df.to_csv("belt_updated_coordinates.csv", index=False)

print("Updated coordinates saved to updated_coordinates.csv")
```

```python
import shutil
import os

def copy_tif_files(source_folder, destination_folder, min_x, max_x, min_y, max_y):
    # Iterate through files in the source folder
    for filename in os.listdir(source_folder):
        if filename.endswith(".tif"):
            # Remove the file extension
            filename_without_extension = os.path.splitext(filename)[0]
            # Assuming the filename follows a pattern where coordinates are included in the filename
            # Extracting coordinates from the filename
            # print(filename_without_extension)
            coordinates = filename_without_extension.split("_")
            x_coord = int(coordinates[0])
            y_coord = int(coordinates[1])
            # print(x_coord, y_coord)

            # Check if coordinates are within the specified range
            if min_x <= x_coord <= max_x and min_y <= y_coord <= max_y:
                # Copy the file to the destination folder
                shutil.copy(os.path.join(source_folder, filename), destination_folder)
                print(f"File {filename} copied.")

# Example usage:
source_folder = "D:\\Yehmh\\test_py\\202301\\P00073_transect_234\\5m_5m\\unknown"
destination_folder = "D:\\Yehmh\\test_py\\species\\Pterocarpus_indicus"
min_x = 300502
max_y = 2631231
max_x = min_x + 100
min_y = max_y - 20


copy_tif_files(source_folder, destination_folder, min_x, max_x, min_y, max_y)
```

```python
import pandas as pd
import shutil

species_mapping = pd.read_csv('D:\\Yehmh\\test_py\\species_percentage.csv')
species_mapping = species_mapping[species_mapping['Percentage'] >= 80]

data_dir = 'D:/Yehmh/test_py/202301/P00073_transect_234/5m_5m/known'
Folder_names = "T"

# Create a dictionary to store folder paths for each species ID
species_folders = {}
# species_folder_path = 'D:\\Yehmh\\test_py\\species\\5m_5m'

for folder in os.listdir(data_dir):
    if folder.startswith(Folder_names):
        species_info = species_mapping.loc[species_mapping["Plot"] == folder]
        if not species_info.empty and not pd.isna(species_info["SpeciesID"].values[0]):
            species = species_info["SpeciesID"].values[0]
            
            # Create a folder for the species ID if it doesn't exist
            if species not in species_folders:
                species_folder_path = os.path.join(data_dir, f"Species_{species}")
                os.makedirs(species_folder_path, exist_ok=True)
                species_folders[species] = species_folder_path
            
            for file in os.listdir(os.path.join(data_dir, folder)):
                if file.endswith(".tif"):
                    image_path = os.path.join(data_dir, folder, file)
                    # Copy the TIF file to the corresponding species folder
                    shutil.copy(image_path, species_folders[species])
        else:
            print(f"Folder {folder} does not have a corresponding species label.")
```
