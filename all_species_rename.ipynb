```python
import os
import pandas as pd

# Path to the directory containing your species files
directory = 'G:\\Yehmh\\DNDF\\tree_survey'

# Initialize an empty list to store all species
all_species = []

# Loop through each file in the directory
for filename in os.listdir(directory):
    if filename.endswith(".xlsx"):  # Assuming all files are Excel files
        file_path = os.path.join(directory, filename)
        # Read the Excel file
        df = pd.read_excel(file_path)
        # Extract the species from the 'Species' column, adjust column name as needed
        species = df['Species'].tolist()
        # Add species to the list
        all_species.extend(species)

# Remove duplicate species
unique_species = list(set(all_species))

# Create a DataFrame with the unique species
species_df = pd.DataFrame({'Species': unique_species})

# Write the DataFrame to an Excel file
output_file = 'G:\\Yehmh\\DNDF\\all_species.xlsx'
species_df.to_excel(output_file, index=False)

print(f"All species have been written to {output_file}")
```

```python
import os
import pandas as pd

# Function to square and sum the 'dbh' values
def square_and_sum_dbh(df):
    return (df['DBH'] ** 2).sum()

# Path to the folders
folder1_path = 'G:\\2024_NDHU research assistant\\DNDF\\tree_survey_coord_00'
folder2_path = 'G:\\2024_NDHU research assistant\\DNDF\\tree_survey'

# Iterate through each file in the first folder
for filename in os.listdir(folder2_path):
    numbers = re.findall(r'\d+', filename)
    if len(numbers) > 2:
        plot, transect = numbers[-2:]  # Extracting the last two numbers

    # Read the CSV file from folder1
    df1 = pd.read_csv(os.path.join(folder1_path, f'{plot}_{transect}.csv'))
    
    # Extract unique tags
    unique_tags = df1['Tag'].unique()
    # print(unique_tags)
    
    # Iterate through unique tags
    for tag in unique_tags:
        # Find corresponding file in folder2
        df2_filename = os.path.join(folder2_path, filename)
        df2 = pd.read_excel(df2_filename)
        
        # Filter rows with matching tag
        matching_rows = df2[df2['Tag'] == tag]
        
        # Calculate squared and summed dbh
        squared_sum = square_and_sum_dbh(matching_rows)
        
        # Append the squared sum to the original file
        df1.loc[df1['Tag'] == tag, 'squared_sum_dbh'] = squared_sum
    
    print(plot, transect)

    # Save the modified CSV file back to folder1
    df1.to_csv(f'G:\\2024_NDHU research assistant\\DNDF\\tree_survey_coord_00_dbh\\{plot}_{transect}.csv', index=False)
```
