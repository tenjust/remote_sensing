```python
import pandas as pd
import twd97

# Read the CSV files into pandas DataFrames
file_path = "D:\\Yehmh\\test_py\\belt_transect_DBH_5up.xlsx"
df = pd.read_excel(file_path)

# Group species data by trap location and find the most common species in each trap
main_species_df = df.groupby(['Plot', 'Transect', 'Quadrat'])['CSP'].agg(lambda x: x.value_counts().idxmax()).reset_index()
# print(main_species_df)

# Calculate the counts of each species in each plot
species_counts = df.groupby(['Plot', 'Transect', 'Quadrat', 'CSP']).size().reset_index(name='Count')
# print(species_counts)

# Calculate the total count of species in each plot
total_counts = species_counts.groupby(['Plot', 'Transect', 'Quadrat'])['Count'].sum().reset_index(name='TotalCount')
# print(total_counts)

# Merge total counts with species counts
species_counts = pd.merge(species_counts, total_counts, on=['Plot', 'Transect', 'Quadrat'], how='left')
# print(species_counts)

# # Calculate the percentage of the main species in each plot
species_counts['Percentage'] = (species_counts['Count'] / species_counts['TotalCount']) * 100
print(species_counts[species_counts['Percentage'] >= 80])

# # Now, you can save the merged DataFrame to a new CSV file if needed
# species_counts.to_csv("species_percentage.csv", index=False)
```

```python
import pandas as pd

# Load the Excel file
df = pd.read_excel("D:\\Yehmh\\test_py\\belt_location.xlsx")

# Function to calculate missing coordinates
def fill_coordinates(row):
    if pd.notna(row['latitude_lu']):
        return row['latitude_lu'], row['longitude_lu']
    elif pd.notna(row['latitude_ld']):
        return row['latitude_ld'] + 20 * 0.000008983, row['longitude_ld']
    elif pd.notna(row['latitude_ru']):
        return row['latitude_ru'], row['longitude_ru'] - 100 * 0.000008983
    elif pd.notna(row['latitude_rd']):
        return row['latitude_rd'] + 20 * 0.000008983, row['longitude_rd'] - 100 * 0.000008983

# Apply the function to fill missing coordinates
df['latitude_lu'], df['longitude_lu'] = zip(*df.apply(fill_coordinates, axis=1))

# Display the updated dataframe
print(df)

# Save the updated dataframe to CSV
df.to_csv("belt_updated_coordinates.csv", index=False)

print("Updated coordinates saved to updated_coordinates.csv")
```
